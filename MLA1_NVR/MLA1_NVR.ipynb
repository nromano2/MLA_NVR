#### **Renaming the Weather Dataset Columns**

```python
df.rename(columns = {'Date/Time':'date_time',
                     'Temp (C)':'temp_c',
                     'Dew Point Temp (C)':'dew_point',
                     'Rel Hum (%)':'relative_humidity',
                     'Wind Spd (km/h)':'wind_speed',
                     'Visibility (km)':'visibility',
                     'Stn Press (kPa)':'pressure',
                     'Weather':'weather'},
          inplace = True)
```

---
## Data Manipulation

### **Adding a Farienhiet Temperature Column `temp_f` to the dataframe**

```python
df['temp_f'] = ((9/5) * df.temp_c) + 32
```

### **Rearranging the Order of the Columns so that `temp_f` preceeds the column `temp_c`**

```python
df = df[['date_time',
         'temp_f',
         'temp_c',
         'dew_point',
         'relative_humidity',
         'wind_speed',
         'visibility',
         'pressure',
         'weather']]
df
```

### **Displaying the datatypes of each of the columns**

```python
df.dtypes
```

### **Changing the `date_time` column datatype to a datatime datatype**

```python
df.date_time = pd.to_datetime(df.date_time)
```

#### *Checking the datatypes after changing the datatype of the `date_time` column*

```python
df.dtypes
```

---
## Data Retrieval

### **Retrieving Monthly Temperature Data**

```python
monthly_avg = pd.Series(df.temp_f.groupby(df.date_time.dt.month).mean())
print(monthly_avg)
```

---
## Data Visualizations

### **Scatterplot of `relative_humidity` against `temp_f`**

```python
scatterplot = plt.figure(figsize=(20,10))
plt.scatter(df.temp_f, df.relative_humidity, color='green')
plt.title('Temperature vs. Relative Humidity')
plt.ylabel('Relative Humidity')
plt.xlabel('Temperature (F)')
```

### **Average Monthly Temperature Lineplot**

```python
Monthly_AVG_lineplot = plt.figure(figsize=(20,10))
plt.plot(monthly_avg, color = 'red')
plt.title('Average Temperature per Month')
plt.ylabel('Temperature (F)')
plt.xlabel('Month')
plt.xticks(list(monthly_avg.index))
```

### **Boxplot of the Numerical Columns of the Weather Dataset**

```python
numerical_columns = [df.temp_f,
                     df.temp_c,
                     df.dew_point,
                     df.relative_humidity,
                     df.wind_speed,
                     df.visibility,
                     df.pressure]

weather_boxplot = plt.figure(figsize=(20,10))
plt.boxplot(numerical_columns)
plt.xticks([1,2,3,4,5,6,7], ['temp_f', 'temp_c', 'dew_point',
              'relative_humidity', 'wind_speed',
              'visibility', 'pressure'])
```
